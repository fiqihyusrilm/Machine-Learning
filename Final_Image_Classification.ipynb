## Membagi data train dan validasi

```python
# import train_test_split dari sklearn
from sklearn.model_selection import train_test_split
#import image generator dari tf keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
import random
from shutil import copyfile

def train_validation_split(source, train, validation, train_ratio):
  total_size = len(os.listdir(source))
  train_size = int(train_ratio * total_size)
  validation_size = total_size - train_size
  randomized = random.sample(os.listdir(source), total_size)
  train_file = randomized[0:train_size]
  validation_file = randomized[train_size:total_size]

  for i in train_file:
    file_i = source + i
    destination = train + i
    copyfile(file_i, destination)

  for i in validation_file:
    file_i = source + i
    destination = validation + i
    copyfile(file_i, destination)

train_ratio = 0.6

source_paper_0 = '/tmp/rockpaperscissors/paper/'
train_paper_0 = '/tmp/rockpaperscissors/train/paper/'
validation_paper_0 = '/tmp/rockpaperscissors/validation/paper/'
train_validation_split(source_paper_0, train_paper_0, validation_paper_0, train_ratio)


source_scissors_1 = '/tmp/rockpaperscissors/scissors/'
train_scissors_1 = '/tmp/rockpaperscissors/train/scissors/'
validation_scissors_1 = '/tmp/rockpaperscissors/validation/scissors/'
train_validation_split(source_scissors_1, train_scissors_1, validation_scissors_1, train_ratio)

source_rock_2 = '/tmp/rockpaperscissors/rock/'
train_rock_2 = '/tmp/rockpaperscissors/train/rock/'
validation_rock_2 = '/tmp/rockpaperscissors/validation/rock/'
train_validation_split(source_rock_2, train_rock_2, validation_rock_2, train_ratio)
```

## Augmentasi Gambar dan Menggunakan Image Generator

```python
# import tensorflow
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
    rescale = 1./255,
    rotation_range = 35,
    horizontal_flip = True,
    shear_range = 0.3,
    fill_mode = 'nearest',
    width_shift_range = 0.2,
    zoom_range = 0.2
)
validation_datagen = ImageDataGenerator(
    rescale = 1./225,
    rotation_range = 35,
    horizontal_flip = True,
    vertical_flip = True,
    shear_range = 0.3,
    fill_mode = 'nearest',
    width_shift_range = 0.2,
    zoom_range = 0.2
)
```

```python
TRAIN_DIR = '/tmp/rockpaperscissors/train/'
train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(150,150),
    batch_size= 32,
    class_mode='categorical',
)

VALIDATION_DIR = '/tmp/rockpaperscissors/validation/'
validation_generator = validation_datagen.flow_from_directory(
    VALIDATION_DIR,
    target_size = (150,150),
    batch_size = 32,
    class_mode = 'categorical',
)
```

## Membangun Jaringan Saraf Tiruan dengan Sequential

```python
# model sequential
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', # karena multi kelas
              optimizer= 'Adam',
              metrics=['accuracy'])
```

```python
# membuat callback untuk menghindari overfitting
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy') > 0.96):
      print("\nAkurasi di atas 96%, training berhenti.")
      self.model.stop_training = True

callbacks = myCallback()
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch = 41, # 1312/32
    epochs = 50,
    validation_data = validation_generator,
    validation_steps = 27, # 876/32
    verbose =2,
    callbacks=[callbacks]
)
```

## Ploting Akurasi dan Loss

```python
# Grafik Loss Train dan Loss validation
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Plot Loss')
plt.ylabel('Value')
plt.xlabel('Epoch')
plt.legend(loc="upper right")
plt.show()
```

```python
# Grafik Akurasi Train dan Akurasi validation
plt.plot(history.history['accuracy'], label = 'Train Accuracy')
plt.plot(history.history['val_accuracy'], label = 'Validation Accuracy')
plt.title('Plot Akurasi')
plt.ylabel('Value')
plt.xlabel('Epoch')
plt.legend(loc='lower right')
plt.show()
```

### Evaluasi model

```python
# menggunakan confussion_matrix
import numpy as np
from sklearn.metrics import confusion_matrix
validation_generator_for_matrix = validation_datagen.flow_from_directory(
    VALIDATION_DIR,
    batch_size=32,
    class_mode='categorical',
    target_size=(150, 150),
    shuffle = False
)

filenames = validation_generator_for_matrix.filenames
nb_samples = len(filenames)

y_predict = model.predict(validation_generator_for_matrix, steps = nb_samples)
y_pred = np.argmax(y_predict, axis=1)
print(confusion_matrix(validation_generator_for_matrix.classes, y_pred))
```
